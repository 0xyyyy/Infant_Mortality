```python
import pandas as pd
```

```python
data = pd.read_csv("h08.csv")
```

```python
med_inc = pd.DataFrame(data.iloc[:,0:2])
med_inc.columns = ["State", "Median Income (2018)"]
med_inc[med_inc["State"]=="Wyoming"]
```

```python
med_inc.drop_duplicates(inplace=True)
med_inc = pd.DataFrame(med_inc.iloc[6:,])
med_inc = med_inc[med_inc["State"] != "D.C."]
med_inc_final = med_inc.drop(med_inc.tail(4).index)
med_inc_final.reset_index()
len(med_inc_final)
```

```python
med_inc_final
```

```python
abbrevs = pd.read_csv("StateAbbreviations.csv")
med_inc_merged = med_inc_final.merge(abbrevs)
med_inc_merged.drop(columns=["Abbrev", "State"], inplace=True)
med_inc_merged.columns = ["Median Income (2018)", "State"]
cols = ["State", "Median Income (2018)"]
med_inc_merged = med_inc_merged[cols]
med_inc_merged.reset_index()
len(med_inc_merged)
```

```python
med_inc_merged.to_csv("MedianIncome2018byState.csv", index=False)
```
